```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs
```

```python
emission = xr.open_dataset('/home/mpim/m300814/microplastic/ncfile_python/plastics_bin10.nc')
display(emission)
print(emission.lat)
```

```python
plt.figure(figsize=[10,5])
ax = plt.axes(projection=ccrs.Robinson())
emission['emission'][0].plot(vmin=0, vmax=1e-16, transform=ccrs.PlateCarree())
ax.coastlines()
plt.show()
```

```python
plt.figure(figsize=[8,4])
ax=plt.axes(projection=ccrs.Robinson())
# emission['emission'][0].plot(vmin=1e-36, vmax=1e-16, transform=ccrs.PlateCarree())
emission['emission'][0].plot(
    norm=mpl.colors.LogNorm(vmin=1e-36, vmax=1e-16),
    transform=ccrs.PlateCarree(),
    cbar_kwargs={'label': "Emission  ($\mu$m)"})
ax.coastlines()
plt.title('10 microns MP Emission')
ax.set_label('Image Colorbar')
plt.show()
```
